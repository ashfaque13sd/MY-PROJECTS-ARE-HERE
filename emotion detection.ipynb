```python
import tensorflow as tf
import streamlit as st
import numpy as np
import cv2

# Load the model
model = tf.keras.models.load_model("face_recognition_trained_model.keras")

# Read the image in color (3 channels)
image = cv2.imread("train_09773_aligned.jpg", cv2.IMREAD_COLOR)

# Resize to match model input (100x100)
image_resized = cv2.resize(image, (100, 100))

# Normalize pixel values
image_normalized = image_resized / 255.0

# Add batch dimension â†’ shape becomes (1, 100, 100, 3)
image_input = np.expand_dims(image_normalized, axis=0)

# Predict
output = model.predict(image_input)
print("Prediction:", output)
```

```python
output[0]
```

```python
import numpy as np

classes = ["surprise","fear","disgust","happy","sad","angry","neutral"]

# Get the index of the highest probability
predicted_index = np.argmax(output)

# Map to emotion label
predicted_emotion = classes[predicted_index]

print("Predicted Emotion:", predicted_emotion)
print("Confidence:", output[0][predicted_index])
```
