```python
df1 = df.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_scale = ['boardtype', 'boardid', 'temp_max', 'temp_min',
       'temp_avg', 'light_max', 'light_min', 'light_avg', 'humidity_min',
       'humidity_max', 'humidity_avg']
scaler = MinMaxScaler()
df1[columns_to_scale] = scaler.fit_transform(df1[columns_to_scale])
```

```python
df1.head()
```

```python
df.head()
```

### Plot distribution of first six features before scaling

```python
numerical_columns = df.columns[2:8]

# Create separate scatter plots for each column
for col in numerical_columns:
    # Create a histogram
    #plt.figure(figsize = (10,6))
    #sns.set(font_scale=1.5)
    print(f"Skewness of {col}",df[col].skew())
    sns.distplot(df[col])
    plt.title(f"Histogram of {col}")
    plt.show()
```

### Plot distribution of first six features after scaling

```python
numerical_columns1 = df1.columns[2:8]

# Create separate scatter plots for each column
for col1 in numerical_columns1:
    # Create a histogram
    #plt.figure(figsize = (10,6))
    #sns.set(font_scale=1.5)
    print(f"Skewness of {col1}",df[col1].skew())
    sns.distplot(df1[col1])
    plt.title(f"Histogram of {col1}")
    plt.show()
```

there is no change in distribution before and after scaling because scaling does not effect the data it just scale the data for machine learning model where models can perform fast computational work.
